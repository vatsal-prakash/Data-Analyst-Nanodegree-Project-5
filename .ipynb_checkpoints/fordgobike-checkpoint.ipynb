## Introduction

### About the Dataset
This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area.
#### The Data
Each trip is anonymized and includes:

- Trip Duration (seconds)
- Start Time and Date
- End Time and Date
- Start Station ID
- Start Station Name
- Start Station Latitude
- Start Station Longitude
- End Station ID
- End Station Name
- End Station Latitude
- End Station Longitude
- Bike ID
- Bike share for all trip
- User Type (Subscriber or Customer – “Subscriber” = Member or “Customer” = Casual)

This dataset will require some data wrangling in order to make it tidy for analysis. For this project, the data selected belongs to the month of *April 2019*. This data was selected because it had the least number of NULL entries and hence requires less time that will be spent during data wrangling as compared to other datasets available for different months. Some feature engineering will also be performed on the data to create new variables that will be used in the analysis and visualizations

### Addressing Issues 

**Issues**
- start_station_id, end_station_id and bike_id are not string types
- start_station_id, start_station_name, end_station_id, end_station_name have some null values
- start_time and end_time is not a datetime object

### Clean

#### Define

Drop rows with NULL values and reset the index 

**Code**

```python
#drop rows using dropna() function
df.dropna(inplace = True)
df.reset_index(inplace = True, drop = True)
```

**Test**

```python
df.isna().any()
```

```python
df.info()
```

Hence we now no longer have any null entries in our dataframe

**Define**

Convert start time and end time columns to datetime objects

**Code**

```python
#convert using to_datatime function
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df.end_time) 
```

**Test**

```python
df[['start_time', 'end_time']].info()
```

```python
df.head()
```

The two columns start_time and end_time are now datetime objects 

**Define**

Convert start_station_id, end_station_id and bike_id to string objects

**Code**

```python
#convert using astype() function
df['start_station_id'] = df['start_station_id'].astype('str')
df['end_station_id'] = df['end_station_id'].astype('str')
df['bike_id'] = df['start_station_id'].astype('str')
```

```python
df[['start_station_id', 'end_station_id' ,'bike_id']].head()
```

```python
#string slicing to remove decimals
df.start_station_id  = df.start_station_id.str[:-2]
df.end_station_id = df.end_station_id.str[:-2]
df.bike_id = df.bike_id.str[:-2]
```

**Test**

```python
df[['start_station_id', 'end_station_id', 'bike_id']].info()
```

Hence the three columns have been successfully been changed to string types

**Define**

Change user_type and bike_share_for_all_trip columns to category types 

**Code**

```python
#convert using astype() function
df.user_type = df.user_type.astype('category')
df.bike_share_for_all_trip = df.bike_share_for_all_trip.astype('category')
```

**Test**

```python
df[['user_type','bike_share_for_all_trip']].info()
```

## Feature Engineering 

For our project we'll extract the date, month and hour values from the start time and end time columns and store these values in new columns

```python
# Adding Month,Date, Day and Hour Column from start_time and end_time columns 

df['start_month'] = df['start_time'].dt.month_name()
df['start_day'] = df['start_time'].dt.day_name()
df['start_date'] = df['start_time'].dt.day
df['start_hour'] = df['start_time'].dt.hour

df['end_month'] = df['end_time'].dt.month_name()
df['end_day'] = df['end_time'].dt.day_name()
df['end_date'] = df['end_time'].dt.day
df['end_hour'] = df['end_time'].dt.hour
```

We'll also create two more columns for duration in minutes and hours from the duration sec column

```python
#divide the duration_min column by 60
df['duration_min'] = df.duration_sec/60
```

```python
#divide the duration_min column by 60 
df['duration_hour'] = df.duration_min/60
```

```python
#rounding duration hour to two decimal places and duration min to zero decimal places 
df = df.round({'duration_min': 0, 'duration_hour': 2})
```

```python
df.head()
```

```python
#changing duration min to an int type using astype()
df.duration_min = df.duration_min.astype('int')
```

```python
df[['duration_min','duration_hour']].info()
```

We'll also add two new columns for longitude and latitude difference

```python
df['longitude_diff'] = df.start_station_longitude - df.end_station_longitude
```

```python
df['latitude_diff'] = df.start_station_latitude - df.end_station_latitude
```

```python
df.head()
```

Let's rearrange the columns for making it easier to read

```python
print(df.columns)
print(df.shape)
```

```python
#rearranging columns
df = df[['bike_id', 'user_type','bike_share_for_all_trip','start_station_id', 'start_station_name', 'start_time', 
         'start_date', 'start_day','start_hour','start_month','end_station_id', 'end_station_name',  'end_time',
         'end_date','end_day','end_hour','end_month' ,'start_station_latitude','end_station_latitude', 'latitude_diff', 
         'start_station_longitude', 'end_station_longitude','longitude_diff', 'duration_sec', 'duration_min', 'duration_hour']]
```

```python
print(df.columns)
print(df.shape)
```

Now our dataframe is ready for analysis

## Exploratory Data Analysis

### Univariate Exploration

Let's create a histogram for the number of trips based on duration in hours 

```python
#get summary statistics of duration_hour
df['duration_hour'].describe()
```

```python
#plot a histogram on normal scale
bins = np.arange(0, df['duration_hour'].max()+0.1, 0.1)
plt.figure(figsize=[10,10])
plt.hist(data = df, x = 'duration_hour', bins = bins)
plt.xlabel('Time for trip (Hours)')
plt.ylabel('No of trips')
plt.title('Total no of trips based on duration in hours')
plt.show()
```

We can see that most trip duration is less than 5 hours. Let's zoom in on the y-axis to see if there are any outliers present in the data

```python
plt.figure(figsize=[10,10])
bins = np.arange(0,df['duration_hour'].max()+0.1, 0.1)
sns.distplot(df['duration_hour'], bins = bins, kde = False);
plt.xlabel('Time for trip (Hours)')
plt.ylabel('No of trips')
plt.title('Total no of trips based on duration in hours')
plt.ylim(0,20);
```

We can see that the data is distributed till the end and no outliers are present and the data is skewed to the right. Now let's plot the data on a log scale 

```python
log_scale  = np.log10(df['duration_hour'])
```

```python
log_scale.describe()
```

```python
# plot a histogram on a log scale 
plt.figure(figsize=[10,10])
sns.set_style('darkgrid')
sns.set_palette('colorblind')
bins = 10** np.arange(-1.7,1.5+0.1, 0.1)
ticks = [0.1,0.3,1,3,10,30]
labels = ['{}'.format(v) for v in ticks]
sns.distplot(df['duration_hour'], bins = bins, kde = False);
plt.xlabel('Time for trip (Hours)')
plt.ylabel('No of trips')
plt.title('Total no of trips based on duration in hours')
plt.xscale('log');
plt.xticks(ticks,labels);
```

**Conclusion**

We get a more normally distributed histogram with a slight skew towards the right and we can see that there are two peaks present in the data with the first one being slightly less than 0.1 hours and the other one being slighlty more than 0.2 hours

#### What are the most popular and the least popular start and stop stations?

Let's plot the top 10 most common start station in descending order

```python
#set base color 
base_color = sns.color_palette()[0]
```

```python
plt.figure(figsize=(10,7));
df_order = df['start_station_name'].value_counts() #count the values in descending order
df_index = df_order.index #select the index i.e the station names
sns.countplot(data = df, y = 'start_station_name',order = df_index[:10], color = base_color ); #create a plot of the first 10 values 
plt.xlabel('Total number of trips')
plt.ylabel('Name of start station')
plt.title('Top 10 most common start stations');
```

Let's also plot top 10 most common stop stations in decreasing order

```python
plt.figure(figsize=(10,7));
df_order = df['end_station_name'].value_counts()
df_index = df_order.index
sns.countplot(data = df, y = 'end_station_name',order = df_index[:10], color = base_color );
plt.xlabel('Total number of trips')
plt.ylabel('Name of stop station')
plt.title('Top 10 most common stop stations');
```

Plotting the top 10 least common start stations

```python
plt.figure(figsize=(10,7));
df_order = df['start_station_name'].value_counts()
df_index = df_order.index
sns.countplot(data = df, y = 'start_station_name',order = df_index[-10:], color = base_color ); #select the last 10 values and plot them
plt.xlabel('Total number of trips')
plt.ylabel('Name of start station')
plt.title('Top 10 least common start stations');
x_labels = np.arange(0, 19, 2)
plt.xticks(x_labels, x_labels);
```

Let's now plot the top 10 least common stop stations

```python
plt.figure(figsize=(10,7));
df_order = df['end_station_name'].value_counts()
df_index = df_order.index
sns.countplot(data = df, y = 'end_station_name',order = df_index[-10:], color = base_color );
plt.xlabel('Total number of trips')
plt.ylabel('Name of stop station')
plt.title('Top 10 least common stop stations');
x_labels = np.arange(0, 19, 2)
plt.xticks(x_labels, x_labels);
```

**Conclusions**

We can see from the above visualizations that the most common start and stop station is *San Francisco Caltrain Station 2* with around 4800 start trips and around 7100 stop trips. *Market St at 10th st* and *San Francisco Ferry Building* are the second and third most common start and stop stations.

Meanwhile *SF Test Station, 17th St at Santa Clara St, Delmas Ave at Virginia St* were the least common start and stop stations with less than 10 rides for the month of April 2019.

### Multivariate Exploration

```python
#create seperate lists for numerical variables and categorical variables 
num_var = ['duration_sec', 'start_station_latitude', 'start_station_longitude', 'end_station_latitude',
                'end_station_longitude', 'start_hour']
cat_var = ['user_type', 'bike_share_for_all_trip']
```

Let's plot a correlation plot of the different numeric variables 

```python
# correlation plot
plt.figure(figsize = [12, 8])
sns.heatmap(df[num_var].corr(), annot = True, fmt = '.2f',
           cmap = 'Pastel2_r', center = 0)

plt.show()
```

From the above plot we can see that there is a high positive or negative correlation between a few variables. These are as follows:- 

**High Positive Correlation**
- start station longitude vs end station longitude
- start station latitude vs end station latitude 

**High Negative Correlation**
- start station latitude vs start station longitude 
- start station longitude vs end station latitude 
- end station longitude vs end station latitude 

Let's create a pairwise grid plot for the above numeric variables. To reduce the computation time and for clarity of the plots, we are going to choose a sample of 1000 

```python
#create pairwise grid plot of 1000 samples
g = sns.PairGrid(data = df.sample(1000), vars = num_var)
g = g.map_diag(plt.hist, bins = 20); #plot histograms on the diagonal axis 
g.map_offdiag(plt.scatter); #scatter plot on off diagonal axis 
```

**Conclusions**

From the above charts, we conclude that there is no linear relationship between any two variables. These variables are divided into segments either on the vertical or horizontal axis
